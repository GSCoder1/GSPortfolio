```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random
```

```python
DATADIR_TRAIN = "C:/Users/Gurpreet/Documents/BME571 - ML in BME/COVID_CNN/Covid19-dataset/train"
DATADIR_TEST = "C:/Users/Gurpreet/Documents/BME571 - ML in BME/COVID_CNN/Covid19-dataset/test"
CATEGORIES = ["Covid", "Normal", "Viral Pneumonia"]
```

```python
IMG_SIZE = 224
x_train = []
for category in CATEGORIES:
    path = os.path.join(DATADIR_TRAIN, category)
    class_type = CATEGORIES.index(category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path, img))
        
        img_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        plt.imshow(img_array)
        
        x_train.append([img_array, class_type])
```

```python
x_test = []
for category in CATEGORIES:
    path = os.path.join(DATADIR_TEST, category)
    class_type = CATEGORIES.index(category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path, img))
        
        img_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        plt.imshow(img_array)
        
        x_test.append([img_array, class_type])
```

```python
random.shuffle(x_train)
random.shuffle(x_test)
```

```python
for sample in x_train[:10]:
    print(sample[1])
```

```python
X_train = []
y_train = []
X_test = []
y_test = []
for features, label in x_train:
    X_train.append(features)
    y_train.append(label)
    
for features, label in x_test:
    X_test.append(features)
    y_test.append(label)
```

```python
np.save('features_train.npy',X_train) #saving
np.save('labels_train.npy',y_train) #saving
np.save('features_test.npy',X_test) #saving
np.save('labels_test.npy',y_test) #saving
```
